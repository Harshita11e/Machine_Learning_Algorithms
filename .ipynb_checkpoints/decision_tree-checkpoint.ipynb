```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("play_tennis.csv")

# Encode categorical values
le = LabelEncoder()
df_encoded = df.apply(le.fit_transform)

X = df_encoded.iloc[:, :-1]
y = df_encoded.iloc[:, -1]

# Train model
model = DecisionTreeClassifier(criterion="entropy", random_state=0)
model.fit(X, y)

# -------- Improved Graphical Tree --------
plt.figure(figsize=(18, 12), dpi=150)   # Bigger size + higher DPI

tree.plot_tree(
    model,
    feature_names=X.columns,
    class_names=["No", "Yes"],
    filled=True,
    rounded=True,      # Rounded boxes
    fontsize=10        # Smaller text to avoid overlap
)

plt.title("Decision Tree for PlayTennis Dataset", fontsize=16)
plt.tight_layout()     # Adjust layout properly
plt.show()
```
