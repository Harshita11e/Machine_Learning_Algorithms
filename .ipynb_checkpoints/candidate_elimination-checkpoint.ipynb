```python
import pandas as pd
import numpy as np

df = pd.read_csv("dataset.csv")

X = np.array(df.iloc[:, :-1])
y = np.array(df.iloc[:, -1])

domains = [np.unique(X[:, i]) for i in range(X.shape[1])]

S = ['0'] * X.shape[1]
G = [['?'] * X.shape[1]]

def is_consistent(h, x):
    for i in range(len(h)):
        if h[i] != '?' and h[i] != x[i]:
            return False
    return True

for i in range(len(X)):
    
    if y[i] == "Yes":   
        G = [g for g in G if is_consistent(g, X[i])]
        
        if S == ['0'] * X.shape[1]:
            S = list(X[i])
        else:
            for j in range(len(S)):
                if S[j] != X[i][j]:
                    S[j] = '?'
    
    else: 
        new_G = []
        
        for g in G:
            if is_consistent(g, X[i]):
                
                for j in range(len(g)):
                    if g[j] == '?':
                        for val in domains[j]:
                            if val != X[i][j]:
                                new_h = g.copy()
                                new_h[j] = val
                                new_G.append(new_h)
            else:
                new_G.append(g)
        
        G = new_G


G_unique = []
for g in G:
    if g not in G_unique:
        G_unique.append(g)

G = G_unique

print("Final Specific Hypothesis (S):")
print(S)

print("\nFinal General Hypotheses (G):")
for g in G:
    print(g)
```
